```python
# Import Needed Packages
from matplotlib import pyplot as plt
import cv2
```

```python
def plt_imshow(title, image):
    # Convert Image From BRG to RGB And Display
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(title)
    plt.grid(False)
    plt.show()
```

```python
# Load Image From Disk
image = cv2.imread("images/30th_birthday.png")

# Display Width, Height, Channels
(h,w,c) = image.shape[:3]
print("Width: {} Pixels".format(w))
print("Height: {} Pixels".format(h))
print("Channels: {}".format(c))
```

```python
# Display Image
plt_imshow("Image", image)
```

```python
# Save Image To Disk
cv2.imwrite("images/NewImageJupyter.jpg",image)
```
