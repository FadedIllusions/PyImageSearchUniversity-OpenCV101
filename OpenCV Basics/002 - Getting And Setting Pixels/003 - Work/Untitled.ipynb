```python
# Import Needed Packages
from matplotlib import pyplot as plt
import cv2
```

```python
def plt_imshow(title, image):
    # Convert Image From BGR Tp RGB And Display
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(title)
    plt.grid(False)
    plt.show()
```

```python
# Load Image, Get Dimensions, Display Image
image = cv2.imread("images/adrian.png")
(h,w) = image.shape[:2]
plt_imshow("Original", image)
```

```python
# Images Are NP Arrays, With Origin At (0,0) At Top-Left
(b,g,r) = image[0,0]
print("Pixel at (0,0) - Red: {}, Green: {}, Blue: {}".format(r,g,b))

# Access Pixel At X=50, Y=20
(b,g,r) = image[20,50]
print("Pixel at (20,50) - Red: {}, Green: {}, Blue: {}".format(r,g,b))

# Update Pixel At (50,20) And Set To Red
image[20,50] = (0,0,255)
(b,g,r) = image[20,50]
print("Pixel at (20,50) - Red: {}, Green: {}, Blue: {}".format(r,g,b))
```

```python
# Compute Center Of Image
(cX, cY) = (w//2, h//2)

# Images Are Numpy Arrays
# Using NP Array Slicing, Get Top-Left Corner Of Image
tl = image[0:cY, 0:cX]
plt_imshow("Top-Left Corner", tl)
```

```python
# In Similar Fashion, Crop Top-Right, Bottom-Right, Bottom-Left Corners
tr = image[0:cY, cX:w]
br = image[cY:h, cX:w]
bl = image[cY:h, 0:cX]

plt_imshow("Top-Right Corner", tr)
plt_imshow("Bottom-Right Corner", br)
plt_imshow("Bottom-Left Corner", bl)
```

```python
# Set TL Corner Of Original Image To Green
image[0:cY, 0:cX] = (0, 255, 0)

# Display Updated Image
plt_imshow("Updated", image)
```

```python
# Also Set BR Corner To Red
image[cY:h, cX:w] = (0, 0, 255)
plt_imshow("Checked Image", image)
```
