```python
# Import Needed Packages
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
def plt_imshow(title, image):
    # Convert Image From BGR To RGB Color Space And Display
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(title)
    plt.grid(False)
    plt.show()
```

```python
# Init 300x300 Px, 3 Channel Image With Black Background
canvas = np.zeros((300,300,3), dtype="uint8")
plt_imshow("Canvas", canvas)
```

```python
# Draw Green Line From TL To BR And Display
green = (0,255,0)
cv2.line(canvas, (0,0), (300,300), green)
plt_imshow("Canvas Green Line", canvas)

# Draw 3 Px Thick Red Line From TR To BL
red = (0,0,255)
cv2.line(canvas, (300,0), (0,300), red, 3)
plt_imshow("Canvas Red Line", canvas)
```

```python
# Draw Green 50x50 Px Square, From (10,10) To (60,60)
cv2.rectangle(canvas, (10,10), (60,60), green)
plt_imshow("Canvas Green Square", canvas)

# Draw Rectangle, Red With 5 Px Thickness
cv2.rectangle(canvas, (50,200), (200,225), red, 5)
plt_imshow("Canvas Red Rectangle", canvas)

# Draw Filled In, Blue Rectangle
blue = (255,0,0)
cv2.rectangle(canvas, (200,50), (225,125), blue, -1)
plt_imshow("Canvas Blue Rectangle", canvas)
```

```python
# Reinit Empty Canvas And Compute Center
canvas = np.zeros((300,300,3), dtype="uint8")
(centerX,centerY) = (canvas.shape[1]//2,canvas.shape[0]//2)
```

```python
# Define Color
white = (255,255,255)

# Loop Over Increasing Radii, From 25 Px 50 To 150 In 25 Px Increments
for r in range(0,175,25):
    # Draw White Circle With Current Radius Size
    cv2.circle(canvas, (centerX,centerY), r, white)
    
# Display
plt_imshow("Canvas White Circles", canvas)
```

```python
# Reinit Empty Canvas
canvas = np.zeros((300,300,3), dtype="uint8")
```

```python
# Draw 25 Random Circles
for i in range(0,25):
    # Randomly Generate Radius Between 50 And 200
    # Generate Random Color
    # Pick Random Point Where Circles Will Be Drawn
    radius = np.random.randint(5, high=200)
    color = np.random.randint(0, high=256, size=(3,)).tolist()
    pt = np.random.randint(0, high=300, size=(2,))
    
    # Draw Circle On Canvas
    cv2.circle(canvas, tuple(pt), radius, color, -1)
    
# Display To Screen
plt_imshow("Canvas Random Circles", canvas)
```
