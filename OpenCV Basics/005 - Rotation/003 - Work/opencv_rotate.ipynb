```python
# Import Needed Packages
import matplotlib.pyplot as plt
import imutils
import cv2
```

```python
# Define Function To Display Images In Jupyter Notebooks
def plt_imshow(title, image):
    # Convert Image From BGR To RGB And Display
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(title)
    plt.grid(False)
    plt.show()
```

```python
# Load Image And Display
image = cv2.imread("images/opencv_logo.png")
plt_imshow("Original", image)
```

```python
# Get Image Dims And Calculate Center
(h,w) = image.shape[:2]
(cX,cY) = (w//2,h//2)
```

```python
# Rotate Image 45 Degs (CCW) Around Center
M = cv2.getRotationMatrix2D((cX,cY), 45, 1.0)
rotated = cv2.warpAffine(image, M, (w,h))
plt_imshow("Rotated 45 Degs CCW", rotated)
```

```python
# Rotate Image 90 Degs (CW) Around Center
M = cv2.getRotationMatrix2D((cX,cY), -90, 1.0)
rotated = cv2.warpAffine(image, M, (w,h))
plt_imshow("Rotated 90 Degs CW", rotated)
```

```python
# Rotate Image About An Arbitrary Point
M = cv2.getRotationMatrix2D((10,50), -45, 1.0)
rotated = cv2.warpAffine(image, M, (w,h))
plt_imshow("Rotated Around Arbitrary Point", rotated)
```

```python
# Use imutils To Rotate Image 180 Degs
rotated = imutils.rotate(image, -180)
plt_imshow("imutils Rotated", rotated)
```

```python
# Rotate Image 33 Degs CCW
# Ensure Image Remains In Viewing Area
rotated = imutils.rotate_bound(image, -33)
plt_imshow("Rotated Without Cropping", rotated)
```
